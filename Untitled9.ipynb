```python
# !pip install ucimlrepo
from ucimlrepo import fetch_ucirepo
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
y = y.map(lambda x: 0 if x == 0 else 1)
y.value_counts()
```

```python
for col in categorical_features:
    print(f"{col}: {X[col].unique()}")
```

```python
X_encoded = pd.get_dummies(X, columns=['cp', 'restecg', 'slope', 'thal'], drop_first=True)
X.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

pca = PCA()
X_pca_full = pca.fit_transform(X_scaled)

explained_var = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(explained_var)+1), explained_var, marker='o', linestyle='--', color='purple')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Explained Variance")
plt.grid(True)
plt.show()

n_components = np.argmax(explained_var >= 0.90) + 1

pca_final = PCA(n_components=n_components)
X_pca = pca_final.fit_transform(X_scaled)


plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="coolwarm", alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA - First Two Components")
plt.show()

```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=42)
rf.fit(X_encoded, y)

# Feature importance scores
rf_importances = pd.Series(rf.feature_importances_, index=X_encoded.columns)
top_rf_features = rf_importances.sort_values(ascending=False).head(10).index

# Sort and plot top 15
plt.figure(figsize=(10,6))
rf_importances.sort_values(ascending=False).head(10).plot(kind='bar', color='purple')
plt.title("Top 10 Features - Random Forest Importance")
plt.xlabel("Features")
plt.ylabel("Importance Score")
plt.show()

print(top_rf_features)
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000, solver="liblinear", random_state=42)

rfe = RFE(estimator=log_reg, n_features_to_select=10)
rfe.fit(X_scaled, y)


selected_features_rfe = X_encoded.columns[rfe.support_]

print("Selected by RFE:")
print(list(selected_features_rfe))
```

```python
from sklearn.feature_selection import chi2, SelectKBest

chi2_selector = SelectKBest(score_func=chi2, k=10)
chi2_selector.fit(X_encoded, y)

selected_features_chi2 = X_encoded.columns[chi2_selector.get_support()]
print("Selected by Chi-Square:", list(selected_features_chi2))
```

```python
final_features = set(top_rf_features) | set(selected_features_rfe) | set(selected_features_chi2)
print("Final Selected Features:", list(final_features))

X_selected = X_encoded[list(final_features)]
print("X_selected shape:", X_selected.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

scaler = StandardScaler()
X_selected_scaled = scaler.fit_transform(X_selected)

X_train, X_test, y_train, y_test = train_test_split(
    X_selected_scaled, y, test_size=0.2, random_state=42, stratify=y
)

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, solver="liblinear", random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "SVM": SVC(random_state=42)
}

plt.figure(figsize=(8,6))

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print(f"\n--- {name} ---")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

    if name != "SVM":
      y_proba = model.predict_proba(X_test)[:,1]
      fpr, tpr, _ = roc_curve(y_test, y_proba)
      auc = roc_auc_score(y_test, y_proba)

      plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.2f})")

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves - All Models")
plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from scipy.cluster.hierarchy import dendrogram, linkage
!pip install kneed
from kneed import KneeLocator

wcss = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_selected_scaled)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, wcss, 'bo-')
plt.xlabel('Number of clusters K')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method for Optimal K')
plt.show()

kneedle = KneeLocator(K_range, wcss, curve='convex', direction='decreasing')
best_k_elbow = kneedle.knee
print(f"Best K from Elbow: {best_k_elbow}")


optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_scaled)


linked = linkage(X_selected_scaled, method='ward')
plt.figure(figsize=(10, 5))
dendrogram(linked, truncate_mode='level', p=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# Agglomerative Clustering with same K
hier_labels = AgglomerativeClustering(n_clusters=optimal_k, linkage='ward').fit_predict(X_scaled)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_selected_scaled)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=kmeans_labels, palette='Set1', s=60)
plt.title('K-Means Clusters')

plt.subplot(1,2,2)
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=hier_labels, palette='Set2', s=60)
plt.title('Hierarchical Clusters')
plt.show()


ari_kmeans = adjusted_rand_score(y, kmeans_labels)
nmi_kmeans = normalized_mutual_info_score(y, kmeans_labels)

ari_hier = adjusted_rand_score(y, hier_labels)
nmi_hier = normalized_mutual_info_score(y, hier_labels)

print(f"K-Means: ARI={ari_kmeans:.3f}, NMI={nmi_kmeans:.3f}")
print(f"Hierarchical: ARI={ari_hier:.3f}, NMI={nmi_hier:.3f}")

# Optional: Cross-tabulation
print("\nK-Means vs Actual Disease:")
print(pd.crosstab(kmeans_labels, y))

print("\nHierarchical vs Actual Disease:")
print(pd.crosstab(hier_labels, y))
```

```python
log_reg_params = {
    "C": [0.01, 0.1, 1, 10, 100],
    "penalty": ["l1", "l2"],
    "solver": ["liblinear"]
}
dt_params = {
    "max_depth": [None, 3, 5, 7, 10],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "criterion": ["gini", "entropy"]
}
rf_params = {
    "n_estimators": [50, 100, 200],
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4],
    "bootstrap": [True, False]
}
svm_params = {
    "C": [0.1, 1, 10, 100],
    "kernel": ["linear", "rbf", "poly"],
    "gamma": ["scale", "auto"]
}

from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

param_grids = {
    "Logistic Regression": log_reg_params,
    "Decision Tree": dt_params,
    "Random Forest": rf_params,
    "SVM": svm_params
}

best_models = {}

for name, model in models.items():
    print(f"\nTuning hyperparameters for {name}...")

    params = param_grids[name]

    if name in ["Logistic Regression", "Decision Tree", "Random Forest"]:
        search = GridSearchCV(model, params, cv=5, scoring="roc_auc", n_jobs=-1)
    else:
        search = RandomizedSearchCV(model, params, cv=5, scoring="roc_auc", n_iter=20, n_jobs=-1, random_state=42)

    search.fit(X_train, y_train)
    best_models[name] = search.best_estimator_

    print(f"Best params for {name}: {search.best_params_}")

plt.figure(figsize=(8,6))

for name, model in best_models.items():
    y_pred = model.predict(X_test)

    print(f"\n--- {name} (Optimized) ---")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))

    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:,1]
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        auc = roc_auc_score(y_test, y_proba)
        plt.plot(fpr, tpr, label=f"{name} (AUC={auc:.2f})")

plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves - Optimized Models")
plt.legend()
plt.show()

```
